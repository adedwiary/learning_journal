Mencari game Baltimore, di kedua tim yang mencetak lebih dari 100 poin.

```python
df[
    (df["_iscopy"] == 0) &
    (df["pts"] > 100) &
    (df["opp_pts"] > 100) &
    (df["team_id"] == "BLB")
]
```

DataFrame memiliki beberapa kolom, sehingga bisa dikelompokkan/grouping. Grouping bertujuan untuk memisah kolom menjadi beberapa group untuk melakukan perhitungan untuk analisis yang lebih baik.

```python
df.head(3)
```

Mengelompokkan data berdasarkan kolom `fran_id`

```python
df.groupby('fran_id')
```

Grouping kolom fran_id dan pts, serta melihat sum

```python
#grouping per kolom dan ada sumnya juga
df.groupby('fran_id')['pts'].sum()
```

Memampilkan data year_id > 2010 dan fran_id = Knicks, tapi di-grouping berdasarkan year_id dan menampilkan sum dari pts

```python
df[
    (df['year_id'] > 2010) &
    (df['fran_id'] == 'Knicks')
].groupby('year_id')['pts'].sum()
```

Memampilkan data year_id > 2010 dan fran_id = Knicks, tapi di-grouping berdasarkan year_id dan game_location, dan menampilkan sum dari pts

```python
df[
    (df['year_id'] > 2010) &
    (df['fran_id'] == 'Knicks')
].groupby(['year_id', 'game_location'])['pts'].sum()
```

Secara default, Pandas akan mengurutkan output pada fungsi `.groupby()`. Jika tidak ingin mengurutkannya, maka berikan `sort=False`. Parameter ini dapat menyebabkan peningkatan kinerja.

```python
df.groupby("fran_id", sort=False)["pts"].sum()
```

## Missing Values

`.info()` menunjukkan berapa banyak nilai non-null yang terdapat dalam sebuah kolom. Ini adalah informasi penting, karena nilai nol sering kali menunjukkan adanya masalah dalam proses pengumpulan data.

Hanya kolom notes yang memiliki nilai null.

Cara termudah untuk menangani missing values adalah dengan mengabaikannya. Kita dapat menghapus semua baris dengan nilai yang hilang menggunakan `.dropna()`:

```python
rows_without_missing_data = nba.dropna()
```

```python
rows_without_missing_data.shape
```

Pembersihan data sepeti ini sangat tidak masuk akal, karena tidak ada masalah jika game kekurangan kolom notes. Tetapi jika dataset berisi satu juta notes yang valid, di mana jika data yang relevan tidak ada, maka menghapus notes yang tidak lengkap bisa menjadi solusi yang masuk akal.

Kita juga dapat menghapus kolom bermasalah jika tidak relevan untuk analisis Anda. Untuk melakukan ini, gunakan .dropna() lagi dan berikan parameter axis=1. Atau menghapus hanya dengan kolomnya saja.

## Invalid Values

Invalid values bisa lebih berbahaya daripada missing values. Seringkali, data yang didapatkan aneh. Ini sangat penting jika dataset sangat besar atau menggunakan entri manual. Nilai yang tidak valid seringkali lebih sulit untuk dideteksi, tetapi kita dapat menerapkan beberapa pemeriksaan dengan kueri dan agregasi.

Satu hal yang dapat dilakukan adalah memvalidasi rentang data. Untuk ini, `.describe()` cukup berguna.

```python
nba.describe()
```

Year_id bervariasi antara 1947 dan 2015. Kedengarannya masuk akal.

Bagaimana dengan points? Apa bisa memiliki nilai minimal 0? Ternyata ada nilai nol di kolom pts, pemain gagal dalam menjalankan gamenya. Namun, tergantung dari analisis, mungkin kita bisa menghapusnya dari data set.

```python
nba[nba["pts"] == 0]
```

**Inconsistent Values**

Terkadang suatu nilai akan sepenuhnya realistis dengan sendirinya, tetapi tidak sesuai dengan nilai di kolom lainnya. Kita dapat menentukan beberapa kriteria kueri yang saling eksklusif dan memverifikasi bahwa ini tidak terjadi bersamaan.

Dalam dataset NBA, nilai dari kolom pts, opp_pts, dan game_result harus konsisten satu sama lain. Kita dapat memeriksanya menggunakan atribut `.empty`:

```python
nba[(nba["pts"] > nba["opp_pts"]) & (nba["game_result"] != 'W')].empty
```

```python
nba[(nba["pts"] < nba["opp_pts"]) & (nba["game_result"] != 'L')].empty
```

Untungnya, kedua queri ini mengembalikan DataFrame kosong.

Bersiaplah untuk kejutan setiap kali Anda bekerja dengan kumpulan data mentah, terutama jika dikumpulkan dari sumber yang berbeda atau melalui alur yang kompleks.
