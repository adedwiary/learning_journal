### Optimisasi

Mencari nilai titik extreme di x sekian

```python
# Optimisasi
## Import pustaka scalar
from scipy.optimize import minimize_scalar

## Mendefinisikan fungsi f dengan python function
def f(x):
  return 2*x**2+4*x-1

## Mendefinisikan optimisasi
opt = minimize_scalar(f)
opt
```

- fun adalah berapa nilai minimumnya
- x adalah posisi nilai ekstrem berupa minimum global

```python
# Mendefinisikan f3 dengan python function
def f3(x):
  return -(x**3 - 6*x**2 + 9*x + 1)

minimize_scalar(f3)
```

Jadi, global maksimum = -5.0 x -1 = 5

## Turunan Parsial

```python
x,y,z = sy.symbols('x y z')
f = 4*x*y + x*sy.sin(z) + x**3 + z**8*y
f
```

```python
#Turunan parsial terhadap x
sy.diff(f,x)
```

```python
#Turunan parsial terhadap y
sy.diff(f,y)
```

```python
f.diff(z)
```

### Gradien

```python
from sympy.tensor.array import derive_by_array
grad = derive_by_array(f, (x,y,z))
grad
```

### Jacobian

```python
f_jacob =sy.Matrix([x**y+y, 2*x*y-2])
f_jacob
```

```python
par = sy.Matrix([x,y])
f_jacob.jacobian(par)
```

```python
sy.Matrix(grad).jacobian(sy.Matrix([x,y,z]))
```

### Hessian

```python
sy.hessian(f,(x,y,z))
```

## Turunan Parsial Numerik

```python
import numpy as np
def f(x,y): 
  return x**2*y+2*x**3*y+y**4

x=np.linspace(1,10)
y=np.linspace(1,10)

z=np.array( [[f(i,j) for i in x] for j in y] )

dx,dy = np.gradient(z)
```

```python
z.shape
```

```python
dx
```

```python
dy
```

```python
#Plot z,dx,dy yang dianggap sebagai sebuah gambar

fig, ax = plt.subplots(ncols=3,figsize=(17,5))
ax[0].imshow(z)
ax[0].set_title('f(x,y)')
ax[1].imshow(dx)
ax[1].set_title('dx')
ax[2].imshow(dy)
ax[2].set_title('dy')
```

## Luas area lingkaran

```python
from scipy.integrate import trapz
import numpy as np

R=30
x=np.linspace(-R,R,200)

upper = np.sqrt(R**2 - x**2)
lower = -upper
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(6,6))
ax.plot(x,upper)
ax.plot(x,lower)
```

```python
area = trapz(upper,x) - trapz(lower,x)

print(area)
```
